# Code to generate the temperature dependencies used in the manuscript titled ""

### These functions describe the relationship between temperature and identified impacts

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Temperature-dependent fish consumption (Kitchell-style)

```python
def x_function(Q10, Tmax, Topt):
    # direct translation of the R expression
    return (((np.log(Q10) * (Tmax - Topt)) ** 2) / 400.0) * (
        1.0 + (1.0 + (40.0 / (np.log(Q10) * (Tmax - Topt + 2.0))) ** 0.5) ** 2
    )

def kitchell_consumption(Tmax, T, Topt, X):
    T = np.asarray(T, dtype=float)
    out = np.zeros_like(T, dtype=float)

    mask = T <= Tmax
    z = (Tmax - T[mask]) / (Tmax - Topt)  # z >= 0 here
    out[mask] = (z**X) * np.exp(X * (1.0 - z))
    return out
```

## 2. Conversion Efficiency (Kitchell-like with 0.57 term)

```python
def x_function_unassim(Q10_unassim, Tmax_unassim, Topt_unassim):
    return (((np.log(Q10_unassim) * (Tmax_unassim - Topt_unassim)) ** 2) / 400.0) * (
        1.0 + (1.0 + (40.0 / (np.log(Q10_unassim) * (Tmax_unassim - Topt_unassim + 2.0))) ** 0.5) ** 2
    )

def kitchell_conversion_eff(Tmax, T, Topt, X):
    T = np.asarray(T, dtype=float)
    out = np.zeros_like(T, dtype=float)

    mask = T <= Tmax
    z = (Tmax - T[mask]) / (Tmax - Topt)  # z >= 0 here
    out[mask] = (z**X) * np.exp(X * (0.57 - z))
    return out
```

## 3. Temperature-dependent Fish Mortality (Arrhenius)


```python
def arrhenius_mort_function(T):
    """
    R: g = 0.001 * exp(10430 * (1/(25+273.15) - 1/(T+273.15)))
    """
    T = np.asarray(T, dtype=float)
    g = 0.001 * np.exp(10430.0 * (1.0 / (25.0 + 273.15) - 1.0 / (T + 273.15)))
    return g
```

## Plot all 3 

```python
# Parameters from your R code
Q10, Tmax, Topt = 2.3, 30.0, 25.0
Q10_unassim, Tmax_unassim, Topt_unassim = 1.5, 30.0, 25.0

X = x_function(Q10, Tmax, Topt)
X_unassim = x_function_unassim(Q10_unassim, Tmax_unassim, Topt_unassim)

# Generate plotting values
t1 = np.arange(0, 42.0 + 1e-9, 0.1)
y1 = kitchell_consumption(Tmax, t1, Topt, X)

t2 = np.arange(0, 42.0 + 1e-9, 0.1)
y2 = kitchell_conversion_eff(Tmax_unassim, t2, Topt_unassim, X_unassim)

t3 = np.arange(0, 40.0 + 1e-9, 0.1)
y3 = arrhenius_mort_function(t3)

# Make the figure
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# A: consumption
axes[0].plot(t1, y1)
axes[0].set_title("A.")
axes[0].set_xlabel("SST (C)")
axes[0].set_ylabel("Fish max. consumption (proportion)")

# B: conversion efficiency
axes[1].plot(t2, y2)
axes[1].set_title("B.")
axes[1].set_xlabel("SST (C)")
axes[1].set_ylabel("Fish conversion efficiency (proportion)")

# C: mortality
axes[2].plot(t3, y3)
axes[2].set_title("C.")
axes[2].set_xlabel("SST (C)")
axes[2].set_ylabel(r"Fish mortality (d$^{-1}$)")

#for ax in axes:
 #   ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
