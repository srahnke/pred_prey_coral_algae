## SST Projections SSP1-2.6

```python
import zipfile, os, glob

zip_path = "ssp_1_2.6.zip"
extract_dir = "sst_extracted_ssp126"   # NEW folder

os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_path, "r") as z:
    z.extractall(extract_dir)

sorted(os.listdir(extract_dir))[:20]
```

```python
import os, glob, re

nc_files = sorted(glob.glob(os.path.join(extract_dir, "**", "*.nc"), recursive=True))
print("Found files:", len(nc_files))
print("First file:", os.path.basename(nc_files[0]))
print("Last file :", os.path.basename(nc_files[-1]))
```

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def preprocess(ds):
    # Drop troublesome aux coords if present
    for v in ["nlat", "nlon"]:
        if v in ds.coords or v in ds.variables:
            ds = ds.drop_vars(v, errors="ignore")
    return ds

ds = xr.open_mfdataset(
    nc_files,
    combine="nested",
    concat_dim="time",
    preprocess=preprocess,
    coords="minimal",
    data_vars="minimal",
    compat="override",
    join="override",
    parallel=False,
)

# Pick SST variable (usually 'tos')
sst_var = "tos" if "tos" in ds.data_vars else list(ds.data_vars)[0]
tos = ds[sst_var]

# You said you already preselected lat/lon when pulling data,
# so we just reduce to a 1D time series by averaging all other dims.
dims_to_mean = [d for d in tos.dims if d != "time"]
ts = tos.mean(dim=dims_to_mean, skipna=True)

# Build a robust monthly time index from cftime objects
tvals = ts["time"].values
year = np.array([t.year for t in tvals], dtype=int)
month = np.array([t.month for t in tvals], dtype=int)
decimal_year = year + (month - 1) / 12.0

# Convert to a pandas Series for convenience
series = pd.Series(ts.values, index=decimal_year).sort_index()

plt.figure()
plt.plot(series.index, series.values)
plt.title(f"{sst_var} time series (mean over {dims_to_mean})")
plt.xlabel("Year (decimal)")
plt.ylabel(f"{sst_var} ({tos.attrs.get('units','')})")
plt.show()
```

### converting output to .csv

```python
out = pd.DataFrame({"decimal_year": series.index, "sst": series.values})
out.to_csv("sst_timeseries_ssp126.csv", index=False)
```

# Combining all Scenarios into one plot

```python
files = {
    "SSP1-2.6": "sst_timeseries_ssp126.csv",
    "SSP2-4.5": "sst_timeseries_ssp245.csv",
    "SSP3-7.5": "sst_timeseries_ssp375.csv",
}

series = {}
for label, path in files.items():
    df = pd.read_csv(path)
    df = df.dropna(subset=["decimal_year", "sst"]).sort_values("decimal_year")
    series[label] = df

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

for ax, (label, df) in zip(axes, series.items()):
    x = df["decimal_year"].values
    y = df["sst"].values
    
    # ---- plot raw series ----
    ax.plot(x, y, label="Monthly SST", color="teal")
    
    # ---- compute linear trend ----
    coeffs = np.polyfit(x, y, 1)     # slope, intercept
    trend = np.polyval(coeffs, x)
    
    # ---- plot trend ----
    ax.plot(x, trend, linestyle="--", label="Trend", color="red")
    
    ax.set_title(label)
    ax.set_xlabel("Year")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="upper left")

axes[0].set_ylabel("SST (C)")
plt.tight_layout()
plt.show()
```

```python
import os, glob, re
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt

# -----------------------------
# USER SETTINGS
# -----------------------------
EXTRACT_DIR = "sst_extracted"   # <-- change to your folder
REGION = "hawaii_box"                 # "hawaii_box" or "global"
LAT_MIN, LAT_MAX = 10, 30
LON_MIN, LON_MAX = 180, 240           # 0360 lon convention (180240E = 180W120W)

# -----------------------------
# 1) Find NetCDF files
# -----------------------------
nc_files = sorted(glob.glob(os.path.join(EXTRACT_DIR, "**", "*.nc"), recursive=True))
if not nc_files:
    raise FileNotFoundError(f"No .nc files found under {EXTRACT_DIR!r}")

# Sort by YYYYMM if present in filenames like *_201501-201912.nc
def extract_yyyymm(path):
    name = os.path.basename(path)
    m = re.search(r"(\d{6})-\d{6}", name)
    return int(m.group(1)) if m else 0

nc_files = sorted(nc_files, key=extract_yyyymm)
print("Found files:", len(nc_files))
print("First file:", nc_files[0])
print("Last file :", nc_files[-1])

# -----------------------------
# 2) Open multifile dataset safely (avoids 'nlat has no index')
# -----------------------------
def preprocess(ds):
    # Drop common troublesome auxiliary coords if they exist (safe)
    for v in ["nlat", "nlon"]:
        if v in ds.coords or v in ds.variables:
            ds = ds.drop_vars(v, errors="ignore")
    return ds

ds = xr.open_mfdataset(
    nc_files,
    combine="nested",
    concat_dim="time",
    preprocess=preprocess,
    coords="minimal",
    data_vars="minimal",
    compat="override",
    join="override",
    parallel=False,
)

# -----------------------------
# 3) Identify SST variable automatically
# -----------------------------
candidates = [v for v in ["tos", "sst", "SST", "temp", "temperature"] if v in ds.data_vars]
if not candidates:
    # fallback: pick the first data_var that has time in its dims
    candidates = [v for v in ds.data_vars if "time" in ds[v].dims]
if not candidates:
    raise ValueError(f"Could not identify an SST variable. Data variables are: {list(ds.data_vars)}")

sst_name = candidates[0]
sst = ds[sst_name]
print("Using variable:", sst_name)
print("Dims:", sst.dims)

# -----------------------------
# 4) Standardize longitude to 0360 if needed
# -----------------------------
if "lon" in sst.coords:
    if float(sst["lon"].min()) < 0:
        sst = sst.assign_coords(lon=((sst["lon"] + 360) % 360)).sortby("lon")

# -----------------------------
# 5) Compute a mean time series (global or Hawaii box)
# -----------------------------
def area_weighted_mean_latlon(da):
    """Area-weighted mean for 1D lat/lon grids."""
    w = np.cos(np.deg2rad(da["lat"]))
    return da.weighted(w).mean(dim=("lat", "lon"), skipna=True)

ts = None

if REGION == "hawaii_box":
    if ("lat" in sst.coords) and ("lon" in sst.coords):
        sst_reg = sst.sel(lat=slice(LAT_MIN, LAT_MAX), lon=slice(LON_MIN, LON_MAX))
        ts = area_weighted_mean_latlon(sst_reg)
    else:
        raise ValueError(
            "Selected REGION='hawaii_box' but dataset does not have 1D lat/lon coords. "
            "If this is a curvilinear grid, we can do a different averaging method."
        )

elif REGION == "global":
    if ("lat" in sst.coords) and ("lon" in sst.coords):
        ts = area_weighted_mean_latlon(sst)
    else:
        # last resort: just average over all non-time dims
        dims = [d for d in sst.dims if d != "time"]
        ts = sst.mean(dim=dims, skipna=True)

else:
    raise ValueError("REGION must be 'hawaii_box' or 'global'")

# -----------------------------
# 6) Convert to pandas time series and plot
# -----------------------------
# Some climate files have cftime times; pandas conversion may fail. Handle both.
try:
    time_index = pd.to_datetime(ts["time"].values)
except Exception:
    # fallback: stringify times
    time_index = pd.to_datetime(ts["time"].astype(str).values)

series = pd.Series(ts.values, index=time_index).sort_index()

plt.figure()
plt.plot(series.index, series.values)
plt.title(f"{sst_name} time series ({REGION})")
plt.xlabel("Time")
plt.ylabel(f"{sst_name} ({sst.attrs.get('units','')})")
plt.show()

series.head(), series.tail()
```
